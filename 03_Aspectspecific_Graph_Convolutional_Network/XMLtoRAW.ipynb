Data Preprocessing for ASGCN


```python
import csv
import requests
import xml.etree.ElementTree as ET
import pandas as pd
import numpy as np

xmlfile="test_or_train.xml"

tree = ET.parse(xmlfile)
#print(tree)
root = tree.getroot()
#print(root)

dataFrame = pd.DataFrame(columns=['sentiment', 'text', 'sentiment_word'])

sentiment = ''
text = ''
sentiment_word = ''
print(len(root.findall('./Document')))
for item in root.findall('./Document'):

  #print(item)
  pos1 = 0
  pos2 = 0





  '''#für modus germeval:
  if not item[0].tag == 'Opinions':
    continue'''
 #modus mobasa:


  for child in item:

    #print(child)
    if child.tag == 'Opinions':
      for p in child:
        sentiment = p.attrib['polarity']
        pos1 = int(p.attrib['from'])
        pos2 = int(p.attrib['to'])
        sentiment_word = text[pos1:(pos2+1)]

        break

    if child.tag == 'text':
      text = child.text


  dataFrame.loc[len(dataFrame.index)] = [sentiment, text, sentiment_word]
print(dataFrame)
```

```python
dataFrame.drop(dataFrame[dataFrame.sentiment_word == ''].index, inplace=True)
dataFrame.reset_index(inplace=True)
dataFrame.drop(columns=['index'], axis=0, inplace=True)
print(dataFrame)
```

```python
f = open("test.raw", "w")
for i in range(len(dataFrame)):
  sentiment = 0
  sentiment_word = dataFrame['sentiment_word'][i]
  text = dataFrame['text'][i].replace(sentiment_word, '$T$', 1)
  if dataFrame['sentiment'][i] == 'positive':
    sentiment = 1
  elif dataFrame['sentiment'][i] == 'neutral':
    sentiment = 0
  elif  dataFrame['sentiment'][i] == 'negative':
    sentiment = -1
  f.write(text + "\n")
  f.write(sentiment_word + "\n")
  f.write(str(sentiment) + "\n")
  #break

f.close()

#open and read the file after the appending:
f = open("test.raw", "r")
print(f.read())
f.close()
```
